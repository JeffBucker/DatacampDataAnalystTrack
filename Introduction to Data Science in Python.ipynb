```python
from matplotlib import pyplot as plt    # Importation du module pyplot de matplotlib
```

```python
plt.plot(df['Instrument_Name'],df['Weight'])    # x 1st param, y 2nd param
plt.plot(df['Instrument_Name'],df['Free_Float'])    # in order to plot a second line 
```

```python
# LABELS

plt.plot(df['Instrument_Name'],df['Weight'])    # x 1st param, y 2nd param
plt.plot(df['Instrument_Name'],df['Free_Float'])    # in order to plot a second line 

plt.xlabel("Sec Name")   # We label the x and y axis
plt.ylabel("Weight")
plt.title("Weight of SPI20 securities")
```

```python
# LEGENDS

plt.plot(df['Instrument_Name'],df['Weight'],label='Index weight')
plt.plot(df['Instrument_Name'],df['Free_Float'],label='Free Float')   

plt.xlabel("Sec Name")
plt.ylabel("Weight")
plt.title("Weight and FFF of SPI20 securities")

plt.legend()    # Use and display the labels previously mentioned

```

```python
# ARBITRARY TEXT AND FONT SIZE

plt.plot(df['Instrument_Name'],df['Weight'],label='Index weight')
plt.plot(df['Instrument_Name'],df['Free_Float'],label='Free Float')   

plt.xlabel("Sec Name", fontsize=10)   # We change the font size of the label
plt.ylabel("Weight")
plt.title("Weight and FFF of SPI20 securities")

plt.text(1, 20,"Degressive")    # we add a text at the specified coordinates

plt.legend() 
```

```python
# ADDING SOME STYLE

plt.plot(df['Instrument_Name'],df['Weight'],label='Index weight',color='seagreen',
         linestyle =':',marker='*', markersize=7.5)    # we change the line color
plt.plot(df['Instrument_Name'],df['Free_Float'],label='Free Float',color='Coral',
         linewidth=3,linestyle='-.')   # we change the linewidth

plt.xlabel("Sec Name", fontsize=10)
plt.ylabel("Weight")
plt.title("Weight and FFF of SPI20 securities")

plt.text(1, 20,"Degressive")

plt.style.use('fivethirtyeight')   # set a grid predefined style

plt.legend() 
```

```python
# Print the list of available styles

print(plt.style.available)
```

```python
# MAKING A SCATTERPLOT

plt.scatter(df['Mcap_Units_local'], df['Weight'], color='green',
            marker='s', alpha=0.5)    # alpha acts on transparency

plt.xlabel("MCap", fontsize=10)
plt.ylabel("Weight")
```

```python
# MAKING A BAR CHART

plt.bar(df['Instrument_Name'],df['Weight'])

plt.xlabel("Sec Name", fontsize=10)
plt.ylabel("Weight")
```

```python
# Display horizontal bars
plt.barh(df['Instrument_Name'],df['Weight'])

plt.ylabel("Sec Name", fontsize=10)
plt.xlabel("Weight")
```

```python
# Stacked bar charts

plt.bar(df['Instrument_Name'],df['Free_Float'], bottom=df['Ci-factor'],label='Sec')
```

```python
# MAKING A HISTOGRAM

# bins = number of category
# range = min and max value
plt.hist(df['Weight'], bins=10, range=(0,20))
```
